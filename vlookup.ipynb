```python
import pandas as pd

sheet1 = pd.DataFrame({
    'EmployeeID': [101, 102, 103, 104],
    'Name': ['Amit', 'Priya', 'Rahul', 'Sneha']
})

sheet2 = pd.DataFrame({
    'EmployeeID': [102, 103, 101, 104],
    'Department': ['IT', 'HR', 'Finance', 'IT'],
    'Salary': [60000, 55000, 70000, 65000]
})

vlookup_result = pd.merge(sheet1, sheet2, on='EmployeeID', how='left')

filtered_result = vlookup_result[
    (vlookup_result['Department'] == 'IT') &
    (vlookup_result['Salary'] >= 60000)
]

print(filtered_result)

filtered_result.to_csv("filtered_data.csv", index=False)
```
