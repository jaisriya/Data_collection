```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame(np.random.rand(100, 14), columns=[f'Col_{i}' for i in range(1, 15)])
df1['primary_key_id'] = range(1, 101)

df2 = pd.DataFrame(np.random.rand(1200, 10), columns=[f'Data_{i}' for i in range(1, 11)])
df2['foreign_key_id'] = np.random.choice(range(1, 101), 1200)
df2['secondary_foreign_key'] = np.random.choice(range(100, 150), 1200)

merged_df = pd.merge(df1, df2, left_on='primary_key_id', right_on='foreign_key_id', how='inner')

print(merged_df.head())
```
