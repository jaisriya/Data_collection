```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print(df.head())

print(df['Fare'].mean())

df_ascending = df.sort_values(by='Fare', ascending=True)
print(df_ascending)

df_descending = df.sort_values(by='Fare', ascending=False)
print(df_descending)

print(df['Fare'].std())
```
